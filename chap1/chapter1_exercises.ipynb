```python
class LogicGate:
    def __init__(self, lbl):
        self.name = lbl
        self.output = None

    def get_label(self):
        return self.name

    def get_output(self):
        self.output = self.perform_gate_logic()
        return self.output


class BinaryGate(LogicGate):
    def __init__(self, lbl):
        super(BinaryGate, self).__init__(lbl)

        self.pin_a = None
        self.pin_b = None

    def get_pin_a(self):
        if self.pin_a == None:
            return int(input("Enter pin A input for gate " + self.get_label() + ": "))
        else:
            return self.pin_a.get_from().get_output()

    def get_pin_b(self):
        if self.pin_b == None:
            return int(input("Enter pin B input for gate " + self.get_label() + ": "))
        else:
            return self.pin_b.get_from().get_output()

    def set_next_pin(self, source):
        if self.pin_a == None:
            self.pin_a = source
        else:
            if self.pin_b == None:
                self.pin_b = source
            else:
                print("Cannot Connect: NO EMPTY PINS on this gate")


class AndGate(BinaryGate):
    def __init__(self, lbl):
        BinaryGate.__init__(self, lbl)

    def perform_gate_logic(self):
        a = self.get_pin_a()
        b = self.get_pin_b()
        if a == 1 and b == 1:
            return 1
        else:
            return 0

class OrGate(BinaryGate):
    def __init__(self, lbl):
        BinaryGate.__init__(self, lbl)

    def perform_gate_logic(self):
        a = self.get_pin_a()
        b = self.get_pin_b()
        if a == 1 or b == 1:
            return 1
        else:
            return 0

class UnaryGate(LogicGate):
    def __init__(self, lbl):
        LogicGate.__init__(self, lbl)

        self.pin = None

    def get_pin(self):
        if self.pin == None:
            return int(input("Enter pin input for gate " + self.get_label() + ": "))
        else:
            return self.pin.get_from().get_output()

    def set_next_pin(self, source):
        if self.pin == None:
            self.pin = source
        else:
            print("Cannot Connect: NO EMPTY PINS on this gate")


class NotGate(UnaryGate):
    def __init__(self, nlbl):
        UnaryGate.__init__(self, lbl)

    def perform_gate_logic(self):
        if self.get_pin():
            return 0
        else:
            return 1


class Connector:
    def __init__(self, fgate, tgate):
        self.from_gate = fgate
        self.to_gate = tgate

        tgate.set_next_pin(self)

    def get_from(self):
        return self.from_gate

    def get_to(self):
        return self.to_gate

# These were left as an exercise
class NorGate(BinaryGate):
    def __init__(self, lbl):
        super().__init__(self, lbl)

    def perform_gate_logic(self):
        a = self.get_pin_a()
        b = self.get_pin_b()

        if a == 0 and b == 0:
            return 1
        else:
            return 0

class NandGate(BinaryGate):
    def __init__(self, lbl):
        super().__init__(self, lbl)

    def perform_gate_logic(self):
        a = self.get_pin_a()
        b = self.get_pin_b()

        if a == 1 and b == 1:
            return 0
        else:
            return 1
```

### 11. The most simple arithmetic circuit is known as the half adder. Research the simple half-adder circuit. Implement this circuit.

```python
class HalfAdder(BinaryGate):
    def __init__(self, lbl):
        super().__init__(lbl)

    def perform_gate_logic(self):
        a = super().get_pin_a()
        b = super().get_pin_b()

        if a == 0 and b == 0:
            return (0, 0)
        elif a == 1 and b == 0:
            return (1, 0)
        elif a == 0 and b == 1:
            return (1, 0)
        else:
            return (0, 1)
```

### 12. Now extend that circuit and implement an 8-bit full adder.

In order to do this, let's first implement a general `TernaryGate` class.

```python
class TernaryGate(LogicGate):
    def __init__(self, lbl):
        super().__init__(lbl)
        self.pin_a = None
        self.pin_b = None
        self.pin_c = None

    def get_pin_a(self):
        if self.pin_a == None:
            return int(input("Enter pin A input for gate " + self.get_label() + ": "))
        else:
            return self.pin_a.get_from().get_output()

    def get_pin_b(self):
        if self.pin_b == None:
            return int(input("Enter pin B input for gate " + self.get_label() + ": "))
        else:
            return self.pin_b.get_from().get_output()

    def get_pin_c(self):
        if self.pin_c == None:
            return int(input("Enter pin C input for gate " + self.get_label() + ": "))
        else:
            return self.pin_b.get_from().get_output()

    def set_next_pin(self, source):
        if self.pin_a == None:
            self.pin_a = source
        else:
            if self.pin_b == None:
                self.pin_b = source
            else:
                if self.pin_c == None:
                    self.pin_b = source
                else:
                    print("Cannot Connect: NO EMPTY PINS on this gate")
```

Now let's implement the `FullAdder` class, changing `get_pin_c()` mehod such that it cannot be changed by the user. We can set this parameter through the method `set_pin_c()` and by default we put it at 0.

Even if it is a bit cumbersome (given that we have now three input variables), we hard code the truth table in the `perform_gate_logic()` method.

```python
class FullAdder(TernaryGate):
    def __init__(self, lbl):
        super().__init__(lbl)
        self.pin_c = 0

    def get_pin_c(self):
        return self.pin_c

    def set_pin_c(self, pin_c):
        self.pin_c = pin_c

    def perform_gate_logic(self):
        a = super().get_pin_a()
        b = super().get_pin_b()
        c = self.get_pin_c()

        if a == 0 and b == 0 and c == 0:
            return (0, 0)
        elif a == 0 and b == 0 and c == 1:
            return (1, 0)
        elif a == 0 and b == 1 and c == 0:
            return (1, 0)
        elif a == 0 and b == 1 and c == 1:
            return (0, 1)
        elif a == 1 and b == 0 and c == 0:
            return (1, 0)
        elif a == 1 and b == 0 and c == 1:
            return (0, 1)
        elif a == 1 and b == 1 and c == 0:
            return (0, 1)
        else:
            return (1, 1)
```

We can now implement a generic __N-BIT FULL ADDER__ by implementing a class which extends the previously created `FullAdder` and using a for-loop to generate the vector of output.

```python
class NBitFullAdder(FullAdder):
    def __init__(self, lbl, n_bit):
        super().__init__(lbl)
        self.n_bit = n_bit

    def perform_gate_logic(self):
        c = 0
        out_vector = []
        
        for _ in range(self.n_bit):
            out, c = super().perform_gate_logic()
            out_vector += [out]

            super().set_pin_c(c)

        return (out_vector, c)
        
full_adder_8_bit = NBitFullAdder("8bfh", 8)
```

We implemented a generic full adder, leaving the user the choice on how many bits to make it.

### 13. The circuit simulation shown in this chapter works in a backward direction. In other words, given a circuit, the output is produced by working back through the input values, which in turn cause other outputs to be queried. This continues until external input lines are found, at which point the user is asked for values. Modify the implementation so that the action is in the forward direction; upon receiving inputs the circuit produces an output.

This will require mainly a rework in how we get the input pins and process.

```python
class LogicGate:
    def __init__(self, lbl):
        self.name = lbl
        self.output = None

    def get_label(self):
        return self.name

    def get_output(self):
        return self.output


class UnaryGate(LogicGate):
    def __init__(self, lbl):
        LogicGate.__init__(self, lbl)

        self.pin = None
        self.output = self.perform_gate_logic()

    def get_pin(self):
        if self.pin == None:
            return int(input("Enter pin input for gate " + self.get_label() + ": "))
        else:
            return self.pin.get_from().get_output()

    def set_next_pin(self, source):
        if self.pin == None:
            self.pin = source
        else:
            print("Cannot Connect: NO EMPTY PINS on this gate")


class BinaryGate(LogicGate):
    def __init__(self, lbl):
        super(BinaryGate, self).__init__(lbl)

        self.pin_a = None
        self.pin_b = None

        self.output = self.perform_gate_logic()

    def get_pin_a(self):
        if self.pin_a == None:
            return int(input("Enter pin A input for gate " + self.get_label() + ": "))
        else:
            return self.pin_a.get_from().get_output()

    def get_pin_b(self):
        if self.pin_b == None:
            return int(input("Enter pin B input for gate " + self.get_label() + ": "))
        else:
            return self.pin_b.get_from().get_output()

    def set_next_pin(self, source):
        if self.pin_a == None:
            self.pin_a = source
        else:
            if self.pin_b == None:
                self.pin_b = source
            else:
                print("Cannot Connect: NO EMPTY PINS on this gate")


class TernaryGate(LogicGate):
    def __init__(self, lbl):
        super().__init__(lbl)
        self.pin_a = None
        self.pin_b = None
        self.pin_c = None

        self.output = self.perform_gate_logic()

    def get_pin_a(self):
        if self.pin_a == None:
            return int(input("Enter pin A input for gate " + self.get_label() + ": "))
        else:
            return self.pin_a.get_from().get_output()

    def get_pin_b(self):
        if self.pin_b == None:
            return int(input("Enter pin B input for gate " + self.get_label() + ": "))
        else:
            return self.pin_b.get_from().get_output()

    def get_pin_c(self):
        if self.pin_c == None:
            return int(input("Enter pin C input for gate " + self.get_label() + ": "))
        else:
            return self.pin_b.get_from().get_output()

    def set_next_pin(self, source):
        if self.pin_a == None:
            self.pin_a = source
        else:
            if self.pin_b == None:
                self.pin_b = source
            else:
                if self.pin_c == None:
                    self.pin_b = source
                else:
                    print("Cannot Connect: NO EMPTY PINS on this gate")
```

By computing the output within the initialization method, we allow the component to produce the output right at the beginning, so that `get_output()` returns the produced value rather than producing the chain of queried outputs. Since the super class `LogicGate` does not have a clue about the pins, we made the modifications in the three direct subclasses.

### 14. Design a class to represent a playing card and another one to represent a deck of cards. Using these two classes, implement your favorite card game.

We will implement the game [__7 e mezzo__](https://en.wikipedia.org/wiki/Sette_e_mezzo) (7 and a half) using [__Sicilian Cards__](https://en.wikipedia.org/wiki/Italian_playing_cards#Spanish-suited_decks). First thing, let's implement the generic Sicilian card.

```python
allowed_suits = ["swords", "cups", "coins", "clubs"]

class SicilianCard:
    def __init__(self, value, suit):
        if suit not in allowed_suits:
            raise Exception("Not a valid suit")
        
        if value < 1 or value > 10:
            raise Exception("Not a valid value")
        
        self.value = value 
        self.suit = suit
        
        if value == 0:
            self.face = "ace"
        elif value == 8:
            self.face = "knight"
        elif value == 9:
            self.face = "queen"
        elif value == 10:
            self.face = "king"
        else:
            self.face = "standard"

    def get_value(self):
        return self.value 

    def get_suit(self):
        return self.suit 

    def get_face(self):
        return self.face if self.face != "standard" else str(self.value)

    def __str__(self):
        return f"{self.get_face().capitalize()} of {self.suit.capitalize()}"

    def __repr__(self):
        return f"SicilianCard({self.get_face().capitalize()} of {self.suit.capitalize()})"
```

Now we can define the class `SicilianCardsDeck`.

```python
import random

class SicilianCardsDeck:
    def __init__(self):
        self.cards = [SicilianCard(i, suit) for i in range(1, 11) for suit in allowed_suits]

    def show_cards(self):
        for card in self.cards:
            print(card)

    def get_size(self):
        return len(self.cards)

    def shuffle(self, n_times=1):
        for _ in range(n_times):
            random.shuffle(self.cards)

    def draw_card(self, shuffle=False):
        if shuffle:
            self.shuffle() 
        return self.cards.pop()

    def put_card(self, card, shuffle=False):
        self.cards += [card]
        if shuffle:
            self.shuffle()

    def reinitialize(self):
        self.__init__()

    def __str__(self):
        return f"Deck of {self.get_size()} Sicilian Cards."
    
    def __repr__(self):
        return f"SicilianCardsDeck({self.get_size()})"
```

Now we can implement the actual game class.

```python
def is_winning_card(card):
    return card.get_suit() == "coins" and card.get_value() == 10

def card_score(card):
    return card.get_value() if card.get_value() <= 7 else 0.5

def hand_value(cards):
    return sum([card_score(card) for card in cards])

class SetteEMezzo:
    def __init__(self, n_of_players):
        self.n_of_players = n_of_players
        self.deck = SicilianCardsDeck()

    def play(self):
        scores = [0] * self.n_of_players

        shuffle_times = int(input("How many times you want to shuffle the deck?"))

        if shuffle_times == 0:
            shuffle_wanted = False 
        else:
            shuffle_wanted = True

        if shuffle_wanted:
            self.deck.shuffle(shuffle_times)

        cards_at_hand = [[self.deck.draw_card()]] * self.n_of_players

        for i, cards in enumerate(cards_at_hand):
            if is_winning_card(cards[0]):
                print(f"Player number {i} won this match because he got {cards[0]}")
                return (i, 7.5, "jolly")

        for current_player in range(self.n_of_players):
            print(f"It's player number {current_player + 1} turn!")
            choice = input("Want to draw?")
            
            while choice != "y" and choice != "n":
                choice = input("Please enter a valid decision!")

            while choice == "y":
                drawn_card = self.deck.draw_card()
                
                if is_winning_card(drawn_card):
                    print(f"Player number {current_player + 1} won with {drawn_card}!")
                    return (current_player, 7.5, "jolly")
                
                cards_at_hand[current_player] += [drawn_card]
                h_val = hand_value(cards_at_hand[current_player])

                if h_val > 7.5:
                    print(f"Player {current_player + 1} lost because he got a hand value of {h_val}!")
                    print(f"Cards at hand for player {current_player + 1} are:")
                    for c in cards_at_hand[current_player]:
                        print(c)
                    cards_at_hand[current_player] = []
                    break

                choice = input("Want to keep drawing?")
            
                while choice != "y" and choice != "n":
                    choice = input("Please enter a valid decision!")

            max_score = 0
            max_index = 0

            for i, hand in enumerate(cards_at_hand):
                if hand != []:
                    scores[i] = 0
                else:
                    scores[i] = hand_value(hand)
                    if scores[i] > max_score:
                        max_score = scores[i]
                        max_index = i

            for i, score in enumerate(scores):
                print(f"Player number {i + 1} made {score}!")

            print(f"Player number {max_index + 1} won with {max_score}!")
            print(f"Cards are:")
            for card in cards_at_hand[max_index]:
                print(card)
           
            return (max_index, max_score, "points")
```
